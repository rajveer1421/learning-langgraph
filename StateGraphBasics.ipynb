### Importing The Required Libraries 


```python
from langgraph.graph import StateGraph,START,END
from  typing import TypedDict 
```

### Defining The State

```python
class OrderState(TypedDict):
    order_id:str
    status:str
    is_inspected:bool
```

### Defining The Nodes 

```python
def recieve_order(state:OrderState):
    print(f" Processing Order:{state['order_id']}")
    return {"status":"Recieved"}
```

```python
def inspect_order(state:OrderState):
    print("Inspecting order")
    return {"status":"Inspected", "is_inspected":True}
```

### Defining The Schema 

```python
builder=StateGraph(OrderState)
builder.add_node("reciever",recieve_order)
builder.add_node("inspector",inspect_order)
```

```python
builder.add_edge(START,"reciever")
builder.add_edge("reciever","inspector")
builder.add_edge("inspector",END)
```

### Compiling The Graph

```python
graph=builder.compile()
```

```python
initial_state=OrderState(order_id="1234",
                       status="Pending",
                       is_inspected=False)
```

```python
print("Starting Execution")
final_state=graph.invoke(initial_state)
print("Graph Execution Finished!")
print("Final State:",final_state)
```
